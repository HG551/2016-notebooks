```python
def read_matrix_to_dictionary(myFile):
    # reads a matrix in a text file
    # ignores # lines
    # returns dictionary with labels of row/column
    inFile = open(myFile,'r')
    header = []
    myDict = {}
    for line in inFile:
        # skip comment lines
        if line[0] == '#':
            continue
        
        line = line.rstrip()
        line = line.split()
        if header == []:
            header = line
#            print 'Header has %i elements' % len(header)
        else:
#            print 'Have a row!, len is %i' % len(line)
#            print line
            rowName = line[0]
            line = line[1:]
            # could also be len(header)
            # a good check would be to make sure header and line are
            # the same length
            for i in range(len(line)):
                colName = header[i]
                value = line[i]
                value = int(value)
                myDict[(rowName,colName)] = value    
    inFile.close()
    return myDict
    
```

```python
d = read_matrix_to_dictionary('BLOSUM62.txt')
```
